Cadastro de Alunos

```python
alunos = []

while True:
    nome = input("Nome do aluno (ou 'sair'): ").strip() #Remove espaços em branco
    if nome.lower() == 'sair': #letras minúsculas
        break

    notas = []
    for i in range(3): #[0,1,2]
        nota = float(input(f"Nota {i+1} de {nome}: "))
        notas.append(nota)

    alunos.append([nome, notas])

print("\n=== MÉDIAS DOS ALUNOS ===")
for aluno in alunos:
    nome, notas = aluno
    media = sum(notas) / len(notas)
    print(f"{nome}: {media:.2f}")

# mostra os alunos acima da média geral
todas_medias = [sum(a[1]) / len(a[1]) for a in alunos]
media_geral = sum(todas_medias) / len(todas_medias)
#[sum(a[1]) / len(a[1]) for a in alunos] - Cria uma lista com todas as médias individuais, usando compreensão de lista

print(f"\nMédia geral da turma: {media_geral:.2f}")
print("Alunos acima da média geral:")
for aluno in alunos:
    media = sum(aluno[1]) / len(aluno[1])
    if media > media_geral:
        print(f" → {aluno[0]} ({media:.2f})")
```

Manipulação de Listas Numéricas

```python
numeros = []
qtd = int(input("Quantos números quer digitar? "))

for i in range(qtd):
    n = int(input(f"Digite o {i+1}º número: "))
    numeros.append(n)

print(f"\nLista original: {numeros}")
print(f"Maior valor: {max(numeros)}")
print(f"Menor valor: {min(numeros)}")

pares = [n for n in numeros if n % 2 == 0]
impares = [n for n in numeros if n % 2 != 0]

print(f"\nNúmeros pares: {pares}")
print(f"Números ímpares: {impares}")

# soma acumulada
soma = 0
for n in numeros:
    soma += n
print(f"\nSoma total: {soma}")
```

Pesquisa de Preços

```python
produtos = []

while True:
    nome = input("Produto (ou 'fim'): ").strip() #retira os espaços vazios
    if nome.lower() == 'fim': #mantém as letras minúsculas
        break
    preco = float(input("Preço: R$ "))
    produtos.append({"nome": nome, "preco": preco})

# ordena por preço
produtos.sort(key=lambda x: x["preco"], reverse=True) #.sort() aceita um argumento chamado key, que indica qual critério usar para comparar os itens durante a ordenação
#lambda é uma função anônima que recebe argumento

print("\n=== PRODUTOS CADASTRADOS ===")
for p in produtos:
    print(f"{p['nome']}: R$ {p['preco']:.2f}")

media_precos = sum([p["preco"] for p in produtos]) / len(produtos)
print(f"\nPreço médio: R$ {media_precos:.2f}")

print("\nProdutos acima da média:")
for p in produtos:
    if p["preco"] > media_precos:
        print(f"→ {p['nome']} (R$ {p['preco']:.2f})")
```

Tabela de Dados

```python
dados = [
    ["Maria", 23, "Engenharia"],
    ["João", 19, "Direito"],
    ["Ana", 21, "Psicologia"],
    ["Carlos", 25, "ADS"]
]

while True:
    nome = input("\nBuscar aluno (ou 'sair'): ").strip()
    if nome.lower() == 'sair':
        break

    encontrado = False
    for pessoa in dados:
        if pessoa[0].lower() == nome.lower():
            print(f"Nome: {pessoa[0]}, Idade: {pessoa[1]}, Curso: {pessoa[2]}")
            encontrado = True
            break

    if not encontrado:
        print("Aluno não encontrado.")
```

Listas com Operações

```python
import random

# gera lista aleatória
lista = [random.randint(1, 100) for _ in range(15)]
print(f"Lista gerada: {lista}")

# filtragens e operações
pares = [n for n in lista if n % 2 == 0]
maiores_50 = [n for n in lista if n > 50]
dobro = [n * 2 for n in lista]
metade = [n / 2 for n in lista]

print(f"\nPares: {pares}")
print(f"Maiores que 50: {maiores_50}")
print(f"Dobro dos valores: {dobro}")
print(f"Metade dos valores: {metade}")

print(f"\nSoma total: {sum(lista)}")
print(f"Média dos valores: {sum(lista) / len(lista):.2f}")
```

Banco de Dados

```python
filmes = [
    ["Inception", 2010, 9.0],
    ["Matrix", 1999, 8.7],
    ["Oppenheimer", 2023, 8.5],
    ["Avatar", 2009, 7.8],
    ["Interestelar", 2014, 8.6],
    ["Tron", 2025, 7.6]

]

# ordenar por nota
filmes.sort(key=lambda x: x[2], reverse=True) #ordena (sort) do maior para o menor (reverse=True)

print("=== FILMES ORDENADOS POR NOTA ===")
for f in filmes:
    print(f"{f[0]} ({f[1]}) — Nota {f[2]}")

# busca por ano
ano = int(input("\nDigite um ano para buscar filmes: "))
filmes_ano = [f for f in filmes if f[1] == ano]

if filmes_ano:
    print(f"Filmes de {ano}:")
    for f in filmes_ano:
        print(f"→ {f[0]} ({f[2]})")
else:
    print("Nenhum filme encontrado para esse ano.")
```

Sistema Controle de Funcionários

```python
funcionarios = []

# ======== CADASTRO DE FUNCIONÁRIOS ========
while True:
    nome = input("\nNome do funcionário (ou 'sair'): ").strip()
    if nome.lower() == 'sair':
        break
    setor = input("Setor: ").strip().title()
    salario = float(input("Salário: R$ "))
    tempo = int(input("Tempo de empresa (anos): "))

    funcionarios.append([nome, setor, salario, tempo])

print("\n✅ Cadastro concluído!")

# ======== MENU DE OPÇÕES ========
while True:
    print("\n=== MENU ===")
    print("1 - Listar todos os funcionários")
    print("2 - Buscar por nome")
    print("3 - Listar por setor")
    print("4 - Estatísticas salariais")
    print("5 - Aplicar aumento por tempo de empresa")
    print("6 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == '1':
        print("\n--- LISTA DE FUNCIONÁRIOS ---")
        for f in funcionarios:
            print(f"Nome: {f[0]}, Setor: {f[1]}, Salário: R${f[2]:.2f}, Tempo: {f[3]} anos")

    elif opcao == '2':
        busca = input("Digite o nome do funcionário: ").strip().lower()
        encontrado = False
        for f in funcionarios:
            if f[0].lower() == busca:
                print(f"Nome: {f[0]}, Setor: {f[1]}, Salário: R${f[2]:.2f}, Tempo: {f[3]} anos")
                encontrado = True
                break
        if not encontrado:
            print("Funcionário não encontrado.")

    elif opcao == '3':
        setor_busca = input("Digite o setor: ").strip().title()
        lista_setor = [f for f in funcionarios if f[1] == setor_busca]
        if lista_setor:
            print(f"\n--- FUNCIONÁRIOS DO SETOR {setor_busca} ---")
            for f in lista_setor:
                print(f"{f[0]} — Salário: R${f[2]:.2f}, Tempo: {f[3]} anos")
        else:
            print("Nenhum funcionário nesse setor.")

    elif opcao == '4':
        if funcionarios:
            salarios = [f[2] for f in funcionarios]
            maior = max(salarios)
            menor = min(salarios)
            media = sum(salarios)/len(salarios)
            print("\n--- ESTATÍSTICAS SALARIAIS ---")
            print(f"Maior salário: R${maior:.2f}")
            print(f"Menor salário: R${menor:.2f}")
            print(f"Média salarial: R${media:.2f}")
            print("Funcionários com mais de 5 anos de empresa:")
            for f in funcionarios:
                if f[3] > 5:
                    print(f" → {f[0]} ({f[3]} anos)")
        else:
            print("Nenhum funcionário cadastrado.")

    elif opcao == '5':
        for i in range(len(funcionarios)):
            nome, setor, salario, tempo = funcionarios[i]
            if tempo > 5:
                aumento = salario * 0.10
                funcionarios[i][2] += aumento
                print(f"Aumento de 10% aplicado a {nome}. Novo salário: R${funcionarios[i][2]:.2f}")
            elif tempo > 2:
                aumento = salario * 0.05
                funcionarios[i][2] += aumento
                print(f"Aumento de 5% aplicado a {nome}. Novo salário: R${funcionarios[i][2]:.2f}")
            else:
                print(f"Nenhum aumento para {nome}.")

    elif opcao == '6':
        print("Saindo do sistema...")
        break

    else:
        print("Opção inválida, tente novamente.")
```
